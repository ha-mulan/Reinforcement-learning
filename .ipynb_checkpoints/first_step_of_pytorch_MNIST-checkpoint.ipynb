```python
X=mnist.data/255
y=mnist.target
y=y.astype(int)
print(type(y[0]))
plt.imshow(X[2].reshape(28,28),cmap='gray')
```

```python
#data loader
import torch
from torch.utils.data import TensorDataset,DataLoader
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/7,random_state=0)

X_train=torch.Tensor(X_train)
X_test=torch.Tensor(X_test)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)

ds_train=TensorDataset(X_train,y_train)
ds_test=TensorDataset(X_test,y_test)

loader_train=DataLoader(ds_train,batch_size=64,shuffle=True)
loader_test=DataLoader(ds_test,batch_size=64,shuffle=False)

```

* dataLoader 구조  
1.데이터를 훈련데이터와 테스트 데이터로 구별  
2.numpy변수를 텐서를 변환  
3.dataset객체생성  
4.dataset객체를 dataLoader 객체로 변환  




```python
# from torch import nn


# model=nn.Sequential()
# model.add_module('fc1',nn.Linear(28*28*1,100))
# model.add_module('relu1',nn.ReLU())
# model.add_module('fc2',nn.Linear(100,100))
# model.add_module('relu2',nn.ReLU())
# model.add_module('fc3',nn.Linear(100,10))
# print(model)
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self,n_in,n_mid,n_out):
        super(Net,self).__init__()
        self.fc1=nn.Linear(n_in,n_mid)
        #self.relu1=nn.ReLU()
        self.fc2=nn.Linear(n_mid,n_mid)
        #self.relu2=nn.ReLU()
        self.fc3=nn.Linear(n_mid,n_out)
    
    def forward(self,x):
        h1=F.relu(self.fc1(x))
        h2=F.relu(self.fc2(h1))
        output=self.fc3(h2)
        return output
model=Net(28*28,100,10)
print(model)
```

```python
from torch import optim


loss_fn=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.01)
```

```python
def train(epoch):
    model.train()
    for data,targets in loader_train:
        optimizer.zero_grad()
        outputs=model(data)
        loss=loss_fn(outputs,targets)
        loss.backward()
        optimizer.step()
    print("epoch{} : 완료".format(epoch))
```

```python
def test():
    model.eval()
    correct=0
    
    with torch.no_grad():
        for data,targets in loader_test:
            print(targets)
            outputs=model(data)
            _,predicted=torch.max(outputs.data,1)
#             print(predicted)
            correct+=predicted.eq(targets.data.view_as(predicted)).sum()
    data_num=len(loader_test.dataset)
    print(correct,data_num)
    print("정확도 {}".format(100*correct/data_num))
```

```python
for i in range(3):
    train(i)
test()
model.eval()

#print("예상결과: {} 정답: {}",format(output,y_test[2]))
```
